```python
import torch
import numpy as np
import pickle
from transformers import RobertaForSequenceClassification, RobertaTokenizerFast, AutoTokenizer
```

```python
tokenizer = RobertaTokenizerFast.from_pretrained('roberta-large')
model = RobertaForSequenceClassification.from_pretrained('../output/roberta-large/checkpoint-2200')
model.to(0)
```

1. Start annotation with Gold response

```python
import pandas as pd

file = pd.read_csv('./GPT_result.csv')
file = file.iloc[:, 1:]

ref = file['real'].tolist()
preds = file['preds'].tolist()
```

```python
len(ref)
```

```python
file
```

```python
inputs_ = tokenizer(ref, truncation=True, padding=True)
```

```python
inputs_dataloader = []
for i in range(len(inputs_['input_ids'])):
    tmp_dic = {}
    tmp_dic['input_ids']= torch.tensor([inputs_['input_ids'][i]])
    tmp_dic['attention_mask']= torch.tensor([inputs_['attention_mask'][i]])
    inputs_dataloader.append(tmp_dic)
```

```python
from fastprogress.fastprogress import master_bar, progress_bar
model.eval()

inputs_labeled_data = []

id2label = model.config.id2label


for batch in progress_bar(inputs_dataloader):

    tmp=[]  
    
    r_input_ids =batch['input_ids'].numpy()[0].tolist()
    tmp.append(tokenizer.decode(r_input_ids, skip_special_tokens=True))
    
    batch = tuple(batch[t].to(0) for t in batch)
    inputs = { 'input_ids' : batch[0], 'attention_mask' : batch[1]}
    outputs = model(**inputs)
    
    tmp.append(np.argmax(outputs[0].detach().cpu().numpy(), axis = 1)[0])
    inputs_labeled_data.append(tmp)
```

```python
label_ = []
for i in range(len(inputs_labeled_data)):
    label_.append(inputs_labeled_data[i][1])

label_
```

```python
file['label'] = label_
```

```python
immed_file = file[['label', 'real', 'preds']]
```

```python
tmp_preds = immed_file['preds'].tolist()
```

```python
for i in range(len(tmp_preds)):
    tmp_preds[i] = tmp_preds[i].replace('<|endoftext|>', '')
```

```python
immed_file['preds'] = tmp_preds 
```

```python
immed_file
```

2. Make RACoLN model inputs by using GPT output

```python
F_list = []
for i in range(len(immed_file)):
    tmp_l = []
    
    tmp_l.append(tmp_preds[i])
    tmp_l.append(label_[i])
    
    F_list.append(tmp_l)
```

```python
len(F_list)
```

```python
for i in range(len(F_list)):
    F_list[i].append(i)
```

```python
import jsonlines
from collections import OrderedDict
import json

with open("infer.jsonl", "w", encoding="utf-8") as f:
    
    for i in range(len(F_list)):
        my_data = OrderedDict()
        my_data["index"] = int(F_list[i][2])
        my_data["X"] = F_list[i][0]
        my_data["Y"] = F_list[i][0]
        my_data["C"] = int(F_list[i][1])

        json.dump(my_data, f, ensure_ascii=False) 
        f.write("\n")
```

3. incorporate RACoLN output with others.

```python
emotion_pred = []

import jsonlines

with jsonlines.open("RA_output/infer.jsonl") as f:
    ct = 20
    for line in f.iter():
        emotion_pred.append(line['X'])
```

```python
f_emotion_pred = []
for i in range(len(emotion_pred)):
  
    if len(tmp_preds[i].split()) == len(emotion_pred[i].split()):
        f_emotion_pred.append(emotion_pred[i])
         
    elif len(tmp_preds[i].split()) == 0:
        f_emotion_pred.append(emotion_pred[i])
            
    elif len(tmp_preds[i].split()) < len(emotion_pred[i].split()):
    
        tp1 = len(tmp_preds[i].split())
        tp = emotion_pred[i].split()[:tp1]
        
        new_sen =""
        
        for i in range(len(tp)):
            new_sen += tp[i] + " "
        new_sen = new_sen[:-1]
        
        f_emotion_pred.append(new_sen)

    else:
        f_emotion_pred.append(emotion_pred[i])
```

```python
len(f_emotion_pred)
```

```python
immed_file['emotion_preds'] = f_emotion_pred
```
